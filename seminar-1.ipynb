```python
#i) a)
p1 = []
p2 = []
p3 = []
for i in range(100):
    x1 = np.random.normal(0,1,25)
    x2 = np.random.normal(0,1,50)
    x3 = np.random.normal(0,1,100)
    p1.append(kstest(x1,'norm', args=(0,1), mode='approx').pvalue)
    p2.append(kstest(x2,'norm', args=(0,1), mode='approx').pvalue)
    p3.append(kstest(x3,'norm', args=(0,1), mode='approx').pvalue)
    
plt.figure()
plt.plot(sorted(p1), np.linspace(0, 1, 100))
plt.plot([0,1], [0,1])

plt.figure()
plt.plot(sorted(p2), np.linspace(0, 1, 100))
plt.plot([0,1], [0,1])

plt.figure()
plt.plot(sorted(p3), np.linspace(0, 1, 100))
plt.plot([0,1], [0,1])
```

```python
#i) д)
def partnorm(size):
    x = np.random.normal(0, 1, size)
    for i in range(size):
        if abs(x[i])<1:
            x[i] = np.random.uniform(-1, 1)
    return x

p1 = []
p2 = []
p3 = []
for i in range(100):
    x1 = partnorm(25)
    x2 = partnorm(50)
    x3 = partnorm(100)
    p1.append(kstest(x1,'norm', args=(0,1), mode='approx').pvalue)
    p2.append(kstest(x2,'norm', args=(0,1), mode='approx').pvalue)
    p3.append(kstest(x3,'norm', args=(0,1), mode='approx').pvalue)
    
plt.figure()
plt.plot(sorted(p1), np.linspace(0, 1, 100))
plt.plot([0,1], [0,1])

plt.figure()
plt.plot(sorted(p2), np.linspace(0, 1, 100))
plt.plot([0,1], [0,1])

plt.figure()
plt.plot(sorted(p3), np.linspace(0, 1, 100))
plt.plot([0,1], [0,1])
```

```python
#a)
mu = 3
sigma = 2
mu_ = 0
sigma_ = 0
p1 = []
p2 = []
p3 = []
for i in range(1000):
    x1 = np.random.normal(mu,sigma,25)
    mu_ = x1.mean()
    sigma_ = np.sqrt(((x1 - mu_)**2).sum()/24)
    p1.append(kstest(x1,'norm', args=(mu_,sigma_), mode='approx').pvalue)
    
    x2 = np.random.normal(mu,sigma,50)
    mu_ = x2.mean()
    sigma_ = np.sqrt(((x2 - mu_)**2).sum()/49)
    p2.append(kstest(x2,'norm', args=(mu_,sigma_), mode='approx').pvalue)
    
    x3 = np.random.normal(mu,sigma,100)
    mu_ = x3.mean()
    sigma_ = np.sqrt(((x3 - mu_)**2).sum()/99)
    p3.append(kstest(x3,'norm', args=(mu_,sigma_), mode='approx').pvalue)
    
plt.figure()
plt.plot(sorted(p1), np.linspace(0, 1, 1000))
plt.plot([0,1], [0,1])

plt.figure()
plt.plot(sorted(p2), np.linspace(0, 1, 1000))
plt.plot([0,1], [0,1])

plt.figure()
plt.plot(sorted(p3), np.linspace(0, 1, 1000))
plt.plot([0,1], [0,1])
```
